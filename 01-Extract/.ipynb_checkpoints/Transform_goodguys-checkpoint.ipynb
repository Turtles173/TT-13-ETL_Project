# Code used to trasform csv to SQL database

```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Extract csv into a dataframe 
goodguys_file = "output/good_guys.csv"
goodguys_df = pd.read_csv(goodguys_file)
goodguys_df.head()
```

```python
# Extract csv into a dataframe 
bing_lee_file = "output/bing_lee.csv"
bing_lee_df = pd.read_csv(bing_lee_file)
bing_lee_df.head()
```

```python
# Dropping model duplicates and setting it to index 
goodguys_df.drop_duplicates("model", inplace=True)
goodguys_df.set_index("model", inplace=True)
goodguys_df.head()
```

```python
# Cleaning up unnecessary columns  
goodguys_transformed =goodguys_df.drop(columns=['Unnamed: 0'])
goodguys_transformed.head()
```

### Create database connection

```python
connection_string = "postgres:Yas.123@localhost:5432/Television_database"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Test connection by confirming tables 
engine.table_names()
```

### Load DataFrames into database

```python
goodguys_transformed.to_sql(name='goodguys', con=engine, if_exists='append', index=True)
```
